```python
# print(os.path.join(img_path +  img.iloc[0, 0]))
for filename in tqdm(os.listdir(images_path)):
#     print(filename)
    if filename.endswith(".jpg"):
        img = cv2.imread(os.path.join(images_path,filename))
        img_crop_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_adapteq = exposure.equalize_adapthist(img_crop_rgb, clip_limit=0.01)
        out_file = (os.path.join(images_path_save , str(filename[0:-4]) + '.jpg'))
#         print(out_file)
        img_adapteq = cv2.convertScaleAbs(img_adapteq, alpha=(255.0))
        img_adapteq = cv2.cvtColor(img_adapteq, cv2.COLOR_BGR2RGB)

        cv2.imwrite(out_file, img_adapteq)

      
    
```
